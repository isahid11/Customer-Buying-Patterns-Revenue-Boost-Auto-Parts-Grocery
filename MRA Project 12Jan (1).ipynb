## Univariate Analysis

### Numerical Columns:

```python
num_columns = ['QUANTITYORDERED', 'PRICEEACH', 'SALES', 'DAYS_SINCE_LASTORDER', 'MSRP']
for col in num_columns:
    sns.histplot(auto_sales[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

### Categorical Columns:

```python
cat_cols = ['STATUS', 'PRODUCTLINE', 'DEALSIZE', 'COUNTRY']
for col in cat_cols:
    plt.figure(figsize=(8, 6))
    sns.countplot(x=col, data=auto_sales)
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
    plt.show()
```

## Bivariate Analysis

## Sales by Product Line:

```python
sns.boxplot(x='PRODUCTLINE', y='SALES', data=auto_sales)
plt.title("Sales Across Product Lines")
plt.xticks(rotation=45)
plt.show()
```

## Relationship between Quantity and Sales:

```python
# Correlation:
quantity_sales_correlation = auto_sales[['QUANTITYORDERED', 'SALES']].corr()
print("Correlation between Quantity Ordered and Sales:", quantity_sales_correlation.loc['QUANTITYORDERED', 'SALES'])

# Visualization:
sns.scatterplot(x='QUANTITYORDERED', y='SALES', hue='DEALSIZE', data=auto_sales)
plt.title("Sales vs Quantity Ordered")
plt.xlabel("Quantity Ordered")
plt.ylabel("Sales")
plt.show()

```

```python

# Bin and analyze sales per quantity range
quantity_bins = pd.cut(auto_sales['QUANTITYORDERED'], bins=[0, 10, 20, 30, 40, 50])
quantity_sales = auto_sales.groupby(quantity_bins)['SALES'].mean()

quantity_sales.plot(kind='bar')
plt.title("Average Sales by Quantity Ordered")
plt.ylabel("Average Sales")
plt.xlabel("Quantity Ordered (Binned)")
plt.show()
```

### Discount Analysis:

```python
auto_sales['Discount'] = auto_sales['MSRP'] - auto_sales['PRICEEACH']
sns.histplot(auto_sales['Discount'], kde=True)
plt.title('Distribution of Discounts')
plt.show()
```

## Time-Series Analysis

```python
auto_sales['Year'] = auto_sales['ORDERDATE'].dt.year
auto_sales['Month'] = auto_sales['ORDERDATE'].dt.month
auto_sales['Quarter'] = auto_sales['ORDERDATE'].dt.quarter
```

```python
monthly_sales = auto_sales.groupby(['Year', 'Month'])['SALES'].sum().reset_index()
sns.lineplot(x='Month', y='SALES', hue='Year', data=monthly_sales)
plt.title("Monthly Sales Trends")
plt.show()
```

```python
quarterly_sales = auto_sales.groupby(['Year', 'Quarter'])['SALES'].sum().reset_index()
sns.barplot(x='Quarter', y='SALES', hue='Year', data=quarterly_sales)
plt.title("Quarterly Sales Trends")
plt.show()
```

```python
yearly_sales = auto_sales.groupby(['Year'])['SALES'].sum().reset_index()
sns.barplot(x='Year', y='SALES', data=yearly_sales)
plt.title("yearly_sales Trends")
plt.show()
```

```python
# Top-selling product lines based on total sales
top_product_lines = auto_sales.groupby('PRODUCTLINE')['SALES'].sum().sort_values(ascending=False)
print(top_product_lines)
```

```python
dealsize_sales = auto_sales.groupby('DEALSIZE')['SALES'].sum().sort_values(ascending=False)
print(dealsize_sales)
```

```python
country_sales = auto_sales.groupby('COUNTRY')['SALES'].sum().sort_values(ascending=False)
print(country_sales)
```

```python
# Monthly sales trend
monthly_sales = auto_sales.groupby(['Year', 'Month'])['SALES'].sum().reset_index()
print(monthly_sales)

# Quarterly sales trend
quarterly_sales = auto_sales.groupby(['Year', 'Quarter'])['SALES'].sum().reset_index()
print(quarterly_sales)
```

```python
# Assuming 'DISCOUNT_PERCENT' is calculated or provided
# Group by discount ranges and sum the sales
discount_sales = auto_sales.groupby(pd.cut(auto_sales['Discount'], bins=[0, 10, 20, 30, 40, 50]))['SALES'].sum()
print(discount_sales)
```

```python
# Correlation between Quantity Ordered and Sales
quantity_sales_correlation = auto_sales[['QUANTITYORDERED', 'SALES']].corr()
print("Correlation between Quantity Ordered and Sales:", quantity_sales_correlation.loc['QUANTITYORDERED', 'SALES'])

# Group by quantity ranges (bins) and calculate average sales
quantity_bins = pd.cut(auto_sales['QUANTITYORDERED'], bins=[0, 10, 20, 30, 40, 50])
quantity_sales = auto_sales.groupby(quantity_bins)['SALES'].mean()
print(quantity_sales)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv(r"C:\Users\ahmed\Downloads\dataset_group.csv")
```

```python
data.head()
```

```python
# Overview of the data
data.info()
```

```python
data.describe()
```

```python
# Check for missing values
print(data.isnull().sum())
```

```python
# Univariate analysis of numerical columns
# Convert 'Date' to datetime format
data['Date'] = pd.to_datetime(data['Date'], dayfirst=True)

# Extract year, month, and day of the week
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['DayOfWeek'] = data['Date'].dt.dayofweek
```

```python
# Number of unique orders
unique_orders = data['Order_id'].nunique()
print(f"Number of unique orders: {unique_orders}")
```

```python
# Count the occurrences of each product
top_products = data['Product'].value_counts().head(10)
print("Top 10 Most Sold Products:")
print(top_products)
```

```python
# Monthly sales trends (if sales data is available)
monthly_sales = data.groupby(['Year', 'Month'])['Order_id'].nunique().reset_index()
monthly_sales.columns = ['Year', 'Month', 'Number_of_Orders']
print(monthly_sales.head())
```

```python

# Weekly sales trends
weekly_sales = data.groupby(['Year', 'DayOfWeek'])['Order_id'].nunique().reset_index()
weekly_sales.columns = ['Year', 'DayOfWeek', 'Number_of_Orders']
print(weekly_sales.head())
```

## Bivariate Analysis

```python
# Number of orders per product
product_order_count = data.groupby('Product')['Order_id'].nunique().reset_index()
product_order_count.columns = ['Product', 'Number_of_Orders']
product_order_count = product_order_count.sort_values(by='Number_of_Orders', ascending=False).head(10)

import seaborn as sns
import matplotlib.pyplot as plt

# Plotting top-selling products
plt.figure(figsize=(10,6))
sns.barplot(x='Product', y='Number_of_Orders', data=product_order_count)
plt.xticks(rotation=45)
plt.title("Top 10 Products by Number of Orders")
plt.show()
```

```python
# Monthly trend of number of orders
monthly_trend = data.groupby(['Year', 'Month'])['Order_id'].nunique().reset_index()
monthly_trend.columns = ['Year', 'Month', 'Number_of_Orders']

plt.figure(figsize=(10,6))
sns.lineplot(x='Month', y='Number_of_Orders', hue='Year', data=monthly_trend)
plt.title("Monthly Order Trends")
plt.show()
```

```python
# Weekly trend of number of orders
weekly_trend = data.groupby(['Year', 'DayOfWeek'])['Order_id'].nunique().reset_index()
weekly_trend.columns = ['Year', 'DayOfWeek', 'Number_of_Orders']

plt.figure(figsize=(10,6))
sns.barplot(x='DayOfWeek', y='Number_of_Orders', hue='Year', data=weekly_trend)
plt.title("Weekly Order Trends")
plt.show()
```

```python
# Identify pairs of products that are frequently purchased together
from itertools import combinations
from collections import Counter

# Get all combinations of products within each order (if order_id can group items bought together)
order_products = data.groupby('Order_id')['Product'].apply(list)

# Get product pairs
product_pairs = [pair for order in order_products for pair in combinations(order, 2)]
pair_counts = Counter(product_pairs)
top_pairs = pair_counts.most_common(10)

print("Top 10 Most Common Product Pairs:")
for pair, count in top_pairs:
    print(f"{pair}: {count} times")
```
